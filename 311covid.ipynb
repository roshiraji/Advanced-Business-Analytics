```python
import pandas as pd
df = pd.read_csv('311_Service_Requests_Afew months before and after the first COVID lockdown.csv', index_col=[0])

df.head()
```

```python
df.shape
```

```python
df[df.isna().any(axis=1)]  # check for NaN ('Not a Number') values 
```

```python
df_filtered = df[['Location Type', 'Incident Address','City','Borough','Y Coordinate (State Plane)','X Coordinate (State Plane)','Created Date',
                        'Closed Date','Agency','Agency Name','Complaint Type','Descriptor','Resolution Action Updated Date']]
```

```python
df_filtered.head()
```

```python
df_filtered.rename(columns={'Location Type':'location_type','Incident Address':'incident_address','City':'city','Borough':'borough',
                                'X Coordinate (State Plane)':'longitude','Y Coordinate (State Plane)':'latitude','Created Date':'created_date',
                                'Closed Date':'closed_date','Agency':'agency','Agency Name':'agency_name','Complaint Type':'complaint_type',
                                'Descriptor':'description','Resolution Action Updated Date':'resolution_date'},inplace=True)

df_filtered.head()
```

```python
df_filtered[df_filtered.isna().any(axis=1)]  # check for NaN ('Not a Number') values 
```

```python
df_filtered.shape
```

```python
df_filtered = df_filtered.dropna(axis=0) # removes rows containing NaN values
df_filtered.shape
```

```python
import datetime
df_filtered['created_date'] = pd.to_datetime(df_filtered['created_date'])
df_filtered['closed_date'] = pd.to_datetime(df_filtered['closed_date'])
df_filtered['resolution_date'] = pd.to_datetime(df_filtered['resolution_date'])
```

```python
df_filtered.info()
```

```python
start_date = pd.Timestamp('2020-02-24')
end_date = pd.Timestamp('2020-03-25')

# Filter rows where created_date is between start and end (inclusive)
df_before_covid = df_filtered[df_filtered['created_date'].between(start_date, end_date)]

df_before_covid.head()
```

```python
import matplotlib.pyplot as plt

# Group by date and count the number of requests per day
requests_per_day_before = df.groupby(df_before_covid['created_date'].dt.date).size()

# Plotting
plt.figure(figsize=(10, 5))
requests_per_day_before.plot(kind='line', marker='o')
plt.title('Number of Service Requests Over Time Before Covid')
plt.xlabel('Date')
plt.ylabel('Number of Requests')
plt.grid(True)
plt.tight_layout()
plt.xticks(rotation=45)
plt.show()
```

```python
covid_start_date = pd.Timestamp('2020-03-26')
covid_end_date = pd.Timestamp('2020-04-27')

# Filter rows where created_date is between start and end (inclusive)
df_after_covid = df_filtered[df_filtered['created_date'].between(covid_start_date, covid_end_date)]

df_after_covid.head()
```

```python
# Group by date and count the number of requests per day
requests_per_day_after = df.groupby(df_after_covid['created_date'].dt.date).size()

# Plotting
plt.figure(figsize=(10, 5))
requests_per_day_after.plot(kind='line', marker='o')
plt.title('Number of Service Requests Over Time After Covid')
plt.xlabel('Date')
plt.ylabel('Number of Requests')
plt.grid(True)
plt.tight_layout()
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot both on the same figure
plt.figure(figsize=(12, 6))
plt.plot(requests_per_day_before.index, requests_per_day_before.values, label='Before COVID', marker='o')
plt.plot(requests_per_day_after.index, requests_per_day_after.values, label='After COVID', marker='x')

# Formatting
plt.title('Service Requests Per Day: Before vs After COVID')
plt.xlabel('Date')
plt.ylabel('Number of Requests')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plotting as histograms
plt.figure(figsize=(12, 6))

plt.hist(requests_per_day_before.values, bins=20, alpha=0.6, label='Before COVID', color='blue')
plt.hist(requests_per_day_after.values, bins=20, alpha=0.6, label='After COVID', color='orange')

plt.title('Distribution of Daily Service Request Counts')
plt.xlabel('Number of Requests per Day')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Get top 10 agencies before and after Covid Lockdown
top_agencies_before = df_before_covid['agency'].value_counts().head(10)
top_agencies_after = df_after_covid['agency'].value_counts().head(10)

#Plot
plt.figure(figsize=(14, 6))

#Before covid
plt.subplot(1, 2, 1)
top_agencies_before.sort_values().plot(kind='barh', color='green')
plt.title('Top 10 Agency (Before COVID Lockdown)')
plt.xticks(rotation=45)
plt.xlabel('Number of Complaints')
plt.ylabel('Complaint Type')

#After covid
plt.subplot(1, 2, 2)
top_agencies_after.sort_values().plot(kind='barh', color='red')
plt.title('Top 10 Agency (After COVID Lockdown)')
plt.xlabel('Number of Complaints')
plt.ylabel('')

plt.tight_layout()
plt.savefig('Top 10 Complaints Covid.png',dpi=400)
plt.show()
```
